```python
#FUNCTION PLAY
import math as math
def circ(r):
    print("Area of circle=", math.pi*r*r)
    print("\nCircumference of a circle=", math.pi*2*r)

def squar(a):
    print("\nArea of a square=",a*a)
    print("\nPerimeter of a Square=",4*a)

leng=float(input("Enter length:"))
circ(leng)
squar(leng)
```

```python
def L2(lis):
    return lis[0],lis[1]

l=[11,2,3,4,4]
print(L2(l))
```

```python
def L2(lis,def_n=2):
    l=[]
    for i in range(0,len(lis)):
        if i<def_n:
            l.append(lis[i])
        else:
            break
    return l

l=[11,2,3,4,4]
print(L2(l))
print(L2(l,4))
```

```python
def Crt_dct(l1,l2):
    if len(l1)!= len(l2):
        return None
    dic={}
    for i in range(len(l1)):
        dic.update({l1[i]:l2[i]})
    return dic


li1=[1,2,3,4,5,6,7]
li2=['a','b','c','d','e','f','g']
li3=[1,2,3]

print(Crt_dct(li1,li2))
print(Crt_dct(li1,li3))
        
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,6,5,4,3,2,1]

f = open("data.txt", "w")
f.write("Rows:\n")
for i in x:
    f.write(f"{i}  ")
f.write("\n")
for i in y:
    f.write(f"{i}  ")

f.write("\nColumns:\n")
for i, j in zip(x,y):
    f.write(f"{i}  {j}\n")
f.close()
```
